```python
import pandas as pd
from google.colab import files
import folium
```

```python
path = '/content/drive/MyDrive/datasets/onlinefoods.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
coordinates_df = pd.concat([df['latitude'], df['longitude']], axis=1)
```

```python
coordinates_df.head()
```

```python
center_latitude = coordinates_df['latitude'].mean()
center_longitude = coordinates_df['longitude'].mean()

map_center = [center_latitude, center_longitude]
mymap = folium.Map(location=map_center, zoom_start=4)

for _, row in coordinates_df.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']]).add_to(mymap)

mymap.save('points_map.html')
```
